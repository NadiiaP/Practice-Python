```python
import requests
url = "https://www.vanityfair.com/style/society/2014/06/monica-lewinsky-humiliation-culture"
r = requests.get(url)
r_html = r.text
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r_html)
print(soup)
```

```python
td = soup.find_all('p')
[i.contents for i in td]
```

```python
for i in range(228,321):
    u = soup.find_all(attrs={"data-reactid": str(i)})
    if u:
        print(u[0].contents)
```

```python
content = [i.contents for i in soup.select('p[data-reactid]')][0:-7]
content
```
